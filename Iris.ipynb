```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
path = "/kaggle/input/iris-flower-dataset/IRIS.csv"
iris_df = pd.read_csv(r"C:\Users\ADMIN\OneDrive\IRIS.csv")
```

```python
X = iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
iris_df['DBSCAN_Cluster'] = dbscan.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(iris_df['sepal_length'], iris_df['sepal_width'], c=iris_df['DBSCAN_Cluster'], cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```
