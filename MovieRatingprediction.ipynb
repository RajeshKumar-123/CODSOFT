```python
import pandas as pd
import zipfile

# Path to the ZIP archive containing the CSV file
zip_file_path = r"C:\Users\ADMIN\OneDrive\Desktop\pythionvsfiles\archive.zip"

# Extract the CSV file from the ZIP archive
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Assuming there's only one CSV file in the ZIP archive
    csv_file_name = zip_ref.namelist()[0]
    zip_ref.extract(csv_file_name, path='.')
    # Read the extracted CSV file into a DataFrame
    df = pd.read_csv(csv_file_name, encoding='latin1')

```
